```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
import scipy.stats as stats
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

import tensorflow as tf
```

```python
# путь к папке, содержащей файлы Excel
excel_folder = r'C:\ann_code\gosniias'

# путь и имя файла, в который будет записан объединенный CSV
output_csv_file = r'C:\ann_code\ml\hw1\data\FlightData.csv'

# Создайте пустой список для хранения DataFrame из Excel файлов
dfs = []

# Получите список файлов Excel в указанной папке
excel_files = [f for f in os.listdir(excel_folder) if f.endswith('.xlsx') or f.endswith('.xls')]

# Пройдите по каждому файлу Excel и считайте его данные в DataFrame
for excel_file in excel_files:
    file_path = os.path.join(excel_folder, excel_file)
    data = pd.read_excel(file_path)  # Чтение данных из файла Excel
    data.rename(columns={data.columns[-1]: "timestamp"}, inplace=True)
    data['timestamp'] = data['timestamp']//1000
    data = data.groupby('timestamp').mean()
    dfs.append(data)

# Объедините все DataFrame из списка в один DataFrame
combined_data = pd.concat(dfs, ignore_index=True)

# Сохраните объединенные данные в CSV файл
combined_data.to_csv(output_csv_file, index=False)
```

```python
# Загружаем данные
df = pd.read_csv('data/FlightData.csv')
```

```python
# названия столбцов, которые в научном институте сказали, что не потребуются. удаляем
del_columns = ['Flight freeze', 'Position freeze', 'Reposition','WOW_L','WOW_R','WOW_N', 'Altitude']

# Используйте метод drop для удаления столбцов
df.drop(del_columns, axis=1, inplace=True)
```

```python
# объединяем вертикальное и горизонтальное отклонения в один показатель
df['Glidepath deviation'] = np.sqrt((df["Vertical dev'n from glidepath"] - df["Horizontal dev'n"])**2)
```

```python
# Распределение (гистограммы для всех фич)

df.hist(bins=20, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Приводим к более нормальному виду
df["Pitch Euler angle"] = abs(df["Pitch Euler angle"])
df["Angle of attack"] = abs(df["Angle of attack"])

df["Pitch Euler angle"] = (df["Pitch Euler angle"])** (1/3)
df["Angle of attack"] = (df["Angle of attack"])** (1/3)

df.loc[df['Angle of attack'] <= 1, 'Angle of attack'] = df['Angle of attack'].median()
```

```python
# Распределение (после нормализации)
normalized_features = ["Pitch Euler angle","Angle of attack"]
df[normalized_features].hist(bins=20, figsize=(12, 4))
plt.tight_layout()
plt.show()
```

# Собираем датасет

```python
# Разбиваем датасет на X и y
X = df.drop(["Glidepath deviation"], axis=1)
y = df["Glidepath deviation"]
```

```python
# Разбиваем данные на train, test, val.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size = 0.5, random_state = 0)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_val = scaler.transform(X_val)
```

```python
X_train.shape, X_test.shape, X_val.shape
```

# Создаем модель

```python
inputs = tf.keras.layers.Input(shape=X_train.shape[1], name='input')
x_1 = tf.keras.layers.Dense(units=16, activation='relu', name='dense_1') (inputs)
#x_2 = tf.keras.layers.Dense(units=32, activation='relu', name='dense_2') (x_1)
output = tf.keras.layers.Dense(units=1, activation='relu', name='output') (x_1)

model = tf.keras.Model(inputs=inputs, outputs=output)
```

```python
model.summary()
```


# Обучение

```python
model.compile(
    optimizer=tf.keras.optimizers.Nadam(learning_rate=0.005), 
    loss=tf.keras.losses.mean_squared_error, 
)
```

```python
history = model.fit(
    x=X_train,
    y=y_train,
    batch_size=X_train.shape[0],
    epochs=700,
    validation_data=(X_val, y_val),
    shuffle=False,
    steps_per_epoch=1,
    validation_steps=1,
    validation_batch_size=X_val.shape[0],
)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

# Оценка

```python
y_pred = model.predict(X_test)
```

```python
print(mean_absolute_error(y_test, y_pred))
```
